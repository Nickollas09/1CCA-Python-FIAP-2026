```python
# estatisticas descritivas das colunas ---->
# exemplo: voltage (tensão, em volts) e global_intensity (corrente, em amperes)

# media aritimedica ----> mean
media_tensao = tabela['Voltage'].mean()
print(f"a media da tensao sera de: {media_tensao:.2f}V")
# mediana ----> median
mediana = tabela['Voltage'].median()
print(f"a mediana da tensao sera de: {mediana:.2f}V")
# desvio padrao ----> standard deviation ----> std
desvio_padrao = tabela['Voltage'].std()
print(f"o desvio padrao sera de: {desvio_padrao:.2f}V")
```

```python
# estatisticas descritivas das colunas ---->
# exemplo: voltage (tensão, em volts) e global_intensity (corrente, em amperes)

# media aritimedica ----> mean
media_tensao = tabela['Global_intensity'].mean()
print(f"a media da tensao sera de: {media_tensao:.2f}V")
# mediana ----> median
mediana = tabela['Global_intensity'].median()
print(f"a mediana da tensao sera de: {mediana:.2f}V")
# desvio padrao ----> standard deviation ----> std
desvio_padrao = tabela['Global_intensity'].std()
print(f"o desvio padrao sera de: {desvio_padrao:.2f}V")
```

```python
# adicionando uma coluna com potencia aparente calculada
# Potencia aparente = Tensao * corrente
tabela['Potencia_aparente'] = tabela['Voltage'] * tabela['Global_intensity']

# Calculando a magnitude da potência aparente (S) usando P (potência ativa) e Q (potência reativa):
# S = sqrt(P^2 + Q^2)
tabela['Potencia_aparente_P_Q'] = (tabela['Global_active_power']**2 + tabela['Global_reactive_power']**2)**0.5
tabela.head()
```
